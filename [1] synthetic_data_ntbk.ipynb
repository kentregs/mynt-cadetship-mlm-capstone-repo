# Synthetic Data Generation

In this notebook, skleearn's datasets package was used to generate a random regression dataset. 1000 samples with seven feature columns were produced using the make_regression method. An arbitrary noise value of 50 was also set along with a random number generation for data creation using the random_state parameter.

---

```python
# Needed for generating classification, regression, and/or clustering datasets
import sklearn.datasets as dt

import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
```

```python
# Generate 1000 samples with 7 feature columns
X, y = dt.make_regression(
    n_samples=1000,
    n_features=7,
    noise=50,
    random_state=42
) 
```

```python
from sklearn.model_selection import train_test_split

# split training_set data into training, validation, and test numpy arrays
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.20, random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.20, random_state=0)
```

```python
print(f'X_train = {X_train.shape}')
print(f'y_train = {y_train.shape}')
print(f'X_val = {X_val.shape}')
print(f'y_val = {y_val.shape}')
print(f'X_test = {X_test.shape}')
print(f'y_test = {y_test.shape}')
```

```python
main_df = pd.DataFrame(data=X)
main_df['target'] = y

train_df = pd.DataFrame(data=X_train)
train_df['target'] = y_train

valid_df = pd.DataFrame(data=X_val)
valid_df['target'] = y_val

test_df = pd.DataFrame(data=X_test)
test_df['target'] = y_test

main_df.info()
```

```python
main_df.describe()
```

### Plot a paiplot to visualize the relationships between all of the columns in the main dataframe.
---

```python
plt.style.use('fivethirtyeight')

sns.pairplot(main_df, height=5)
plt.show()
```

```python
# export split generated synthetic data as csv files
main_df.to_csv('data/main_data.csv', index=False)
train_df.to_csv('data/training_data.csv', index=False)
valid_df.to_csv('data/validation_data.csv', index=False)
test_df.to_csv('data/test_data.csv', index=False)
```
